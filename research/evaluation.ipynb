```python
from datasets import Dataset

data_local = {
    "question": [item["question"] for item in results],
    "answer": [item["local_answer"] for item in results],
    "contexts": [item["contexts"] for item in results],
    "ground_truth": [item["ground_truth"] for item in results]
}

dataset_local = Dataset.from_dict(data_local)
```

```python
data_open = {
    "question": [item["question"] for item in results],
    "answer": [item["openrouter_answer"] for item in results],
    "contexts": [item["contexts"] for item in results],
    "ground_truth": [item["ground_truth"] for item in results]
}

dataset_open = Dataset.from_dict(data_open)
```

```python
from ragas import evaluate
from ragas.metrics import (
    faithfulness,
    answer_relevancy,
    context_precision,
    context_recall
)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
print(embeddings.embed_query("Hello"))
```

```python
from ragas.embeddings import LangchainEmbeddingsWrapper

evaluator_embeddings= LangchainEmbeddingsWrapper(embeddings)
```

```python
from langchain_openai import ChatOpenAI

evaluator_llm = ChatOpenAI(
    openai_api_base="https://openrouter.ai/api/v1",
    openai_api_key="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",
    temperature=0,
    model="meta-llama/llama-3.1-8b-instruct",
    default_headers={
        "HTTP-Referer": "http://localhost",   # required by OpenRouter
        "X-Title": "rag-eval"
    }
)
print(evaluator_llm.invoke("Hello, evaluator!"))
```

```python
results_local = evaluate(
    dataset_local,
    metrics=[faithfulness, answer_relevancy, context_precision, context_recall],
    llm=evaluator_llm,
    embeddings=embeddings
)
```

```python
results_local
```

```python
results_open = evaluate(
    dataset_open,
    metrics=[faithfulness, answer_relevancy, context_precision, context_recall],
    llm=evaluator_llm,
    embeddings=embeddings
)
```
